```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

invloss_trainingweight_files = ['TOY_invloss_trainingweight_trainset_size_10.pkl',
                                'TOY_DAC_invloss_tau_10000.pkl',
                                'TOY_invloss_trainingweight_trainset_size_100.pkl',
                                'TOY_invloss_trainingweight_trainset_size_200.pkl']

l2_trainingweight_files = ['TOY_l2_trainingweight_trainset_size_10.pkl',
                            'TOY_DAC_l2_tau_19.pkl',
                            'TOY_l2_trainingweight_trainset_size_100.pkl',
                            'TOY_l2_trainingweight_trainset_size_200.pkl']

cosine_trainingweight_files = ['TOY_cosine_trainingweight_trainset_size_10.pkl',
                                'TOY_DAC_cosine_tau_140.pkl',
                                'TOY_cosine_trainingweight_trainset_size_100.pkl',
                                'TOY_cosine_trainingweight_trainset_size_200.pkl']

cosine_origin_trainingweight_files = ['TOY_cosine_origin_trainingweight_trainset_size_10.pkl',
                                        'TOY_DAC_cosine_origin_tau_140.pkl',
                                        'TOY_cosine_origin_trainingweight_trainset_size_100.pkl',
                                        'TOY_cosine_origin_trainingweight_trainset_size_200.pkl']

invloss_priorweight_files = ['TOY_invloss_priorweight_trainset_size_10.pkl',
                                'TOY_DAC_priorweight_invloss_tau_5000.pkl',
                                'TOY_invloss_priorweight_trainset_size_100.pkl',
                                'TOY_invloss_priorweight_trainset_size_200.pkl']

l2_priorweight_files = ['TOY_l2_priorweight_trainset_size_10.pkl',
                            'TOY_DAC_priorweight_l2_tau_19.pkl',
                            'TOY_l2_priorweight_trainset_size_100.pkl',
                            'TOY_l2_priorweight_trainset_size_200.pkl']

cosine_priorweight_files = ['TOY_cosine_priorweight_trainset_size_10.pkl',
                                'TOY_DAC_priorweight_cosine_tau_140.pkl',
                                'TOY_cosine_priorweight_trainset_size_100.pkl',
                                'TOY_cosine_priorweight_trainset_size_200.pkl']

cosine_origin_priorweight_files = ['TOY_cosine_origin_priorweight_trainset_size_10.pkl',
                                        'TOY_DAC_priorweight_cosine_origin_tau_140.pkl',
                                        'TOY_cosine_origin_priorweight_trainset_size_100.pkl',
                                        'TOY_cosine_origin_priorweight_trainset_size_200.pkl']


```

```python
def add_to_plot(files, label):
    
    # load each file
    dfs = []
    for file in files:
        df = pd.read_pickle('test_accuracies/'+file)
        dfs.append(df)

    trainset_sizes = [10, 50, 100, 200]
    means = []
    stds = []
    for df in dfs:
        means.append(df['mean_all'])
        stds.append(df['std_all'])

    # remove first index
    #means = means[1:]
    #stds = stds[1:]
    #trainset_sizes = trainset_sizes[1:]

    # plot means against trainset sizes, with stds as error bars
    #plt.errorbar(trainset_sizes, means, yerr=stds, fmt='o', label=label)
    # a line plot instead
    plt.plot(trainset_sizes, means, 'o-', label=label)
    # add the stds as shaded regions
```

```python
plt.figure()
add_to_plot(invloss_trainingweight_files, 'Inv loss')
add_to_plot(cosine_trainingweight_files, 'Cos grad')
add_to_plot(cosine_origin_trainingweight_files, 'Cos weight')
add_to_plot(l2_trainingweight_files, 'L2')
plt.xlabel('Trainset size')
plt.ylabel('Test loss')
# set y-axis to log scale
plt.yscale('log')
plt.legend()
# save figure
plt.savefig('figs/trainsetsize_fedavg.png')
plt.show()


plt.figure()
add_to_plot(invloss_priorweight_files, 'Inv loss')
add_to_plot(cosine_priorweight_files, 'Cos grad')
add_to_plot(cosine_origin_priorweight_files, 'Cos weight')
add_to_plot(l2_priorweight_files, 'L2')
plt.xlabel('Trainset size')
plt.ylabel('Test loss')
# set y-axis to log scale
plt.yscale('log')
plt.legend()
# save figure
plt.savefig('figs/trainsetsize_fedsim.png')

plt.show()
```
