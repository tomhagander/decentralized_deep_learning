```python
# load save/label_trainingweight_results.pkl

import pickle
import numpy as np
import matplotlib.pyplot as plt

# use seaborn
import seaborn as sns
sns.set()


```

```python
def plot_results(resultsfile):

    with open(resultsfile, 'rb') as f:
        results = pickle.load(f)

    for i, result in enumerate(results):
        # order result based on the value of the first index in each tuple
        result = sorted(result, key=lambda x: x[0])
        # remove all tuples that are (0, 0)
        result = [x for x in result if x != (0, 0)]
        print(result)
        # result is a list of tuples of (x,y) values, plot these as points
        x, y = zip(*result)

        plt.plot(x, y, marker='x', linestyle='-')
        # add title and axis labels
        plt.title('Validation accuracy vs Tau: ' + resultsfile)
        plt.xlabel('Tau')
        plt.ylabel('Average validation accuracy')
        # add legend
        plt.legend(['invloss', 'l2', 'cosine', 'origin'])
        # add
        # make x axis logarithmic
        plt.xscale('log')
    if resultsfile == 'label_trainingweight_results.pkl' or resultsfile == 'label_priorweight_results.pkl':
        # plot a dotted horizontal line with the text oracle at 70.35
        plt.axhline(y=70.35, color='r', linestyle='--', label='oracle')
        plt.axhline(y=68.2, color='y', linestyle='--', label='random')
        plt.legend(['invloss', 'l2', 'cosine', 'origin', 'oracle', 'random'])
    elif resultsfile == 'fivecluster_priorweight_results.pkl' or resultsfile == 'fivecluster_trainingweight_results.pkl':
        plt.axhline(y=91.57, color='r', linestyle='--', label='oracle')
        plt.axhline(y=87.19, color='y', linestyle='--', label='random')
        plt.legend(['invloss', 'l2', 'cosine', 'origin', 'oracle', 'random'])
    elif resultsfile == 'mnist_priorweight_results.pkl' or resultsfile == 'mnist_trainingweight_results.pkl':
        plt.axhline(y=89.64, color='r', linestyle='--', label='oracle')
        plt.axhline(y=82.78, color='y', linestyle='--', label='random')
        plt.legend(['invloss', 'l2', 'cosine', 'origin', 'oracle', 'random'])
    plt.show()
```

```python
resultsfiles = ['label_trainingweight_results.pkl', 'fivecluster_trainingweight_results.pkl', 'fivecluster_priorweight_results.pkl',
                 'label_priorweight_results.pkl', 'mnist_priorweight_results.pkl', 'mnist_trainingweight_results.pkl']

for resultsfile in resultsfiles:
    plot_results(resultsfile)
```
